# Notebook to plot anomaly built

## import module

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import cartopy
import cartopy.crs as ccrs
```

## parameters

```python
# data dir path
cdir='./'

# list of file variables to process and the corresponding infile variables
fvarlst=['huss','tprecip','prsn','psl','rlds','rsds','tas','uas','vas']
cvarlst=['huss','tprecip','prsn','psl','rlds','rsds','tas','uas','vas']

# reference period
yrefs=1975; yrefe=2014

# trg period
ytrgs=2260; ytrge=2299

cfext = '{:d}{:d}-{:d}{:d}'.format(ytrgs,ytrge,yrefs,yrefe)

```

## Process data

```python
# define figure
fig=plt.figure(figsize=(np.array([400,120*6]) / 25.4))
grid = plt.GridSpec(len(fvarlst), 3, wspace=0.2, hspace=0.3)

# process all data
for ivar, fvar in enumerate(fvarlst):
    cvar=cvarlst[ivar]
    
    # load data
    print('    load anomaly data ...')
    cfile_pert='CLIMATO_FORCING/JRA55_ano_{}_monthly_{}_y1951.nc'.format(cfext,fvar)
    cfile_ref='CLIMATO_FORCING/JRA55_monthly_{}_y1951.nc'.format(fvar)
    print('        file : {}'.format(cfile_pert))
    dspert_month = xr.open_dataset(cfile_pert).sel(lat=slice(-90,-50))
    print('        file : {}'.format(cfile_ref))
    dsref_month = xr.open_dataset(cfile_ref).sel(lat=slice(-90,-50))
    
    # compute anual mean
    dspert_annual=dspert_month.mean(dim='time')
    dsref_annual=dsref_month.mean(dim='time')
   
    # compute global mean time series
    # compute weights
    weights = np.cos(np.deg2rad(dspert_month.lat,dtype=np.float64))
    
    # compute weighted xarray
    weights.name = "weights"
    dspertw = xr.open_dataset(cfile_pert).sel(lat=slice(-80,-60)).weighted(weights)
    dsrefw = xr.open_dataset(cfile_ref).sel(lat=slice(-80,-60)).weighted(weights)
    
    # display anomaly
    dspert_ts=dspertw.mean(("lon", "lat"))
    dsref_ts=dsrefw.mean(("lon", "lat"))

    dsano_annual=dspert_annual-dsref_annual
    
    # plot
    vmin=np.min(dsano_annual[cvar].values)
    vmax=np.max(dsano_annual[cvar].values)

    #ax = fig.add_subplot(grid[ivar,0:2], projection=ccrs.Robinson(central_longitude=0))
    ax = fig.add_subplot(grid[ivar,0], projection=ccrs.Stereographic(central_latitude=-90.0, central_longitude=0.0))
    pcol=ax.pcolormesh(dspert_annual['lon'],dspert_annual['lat'],dsano_annual[cvar],vmax=vmax,vmin=vmin,transform=ccrs.PlateCarree())
    ax.contour(dspert_annual['lon'],dspert_annual['lat'],dsano_annual[cvar],levels=[-999999,0,999999],colors='w',transform=ccrs.PlateCarree())
    ax.coastlines()
    XY_lim=[(-180, 180, -90, -60),ccrs.PlateCarree()]
    ax.set_extent(XY_lim[0], XY_lim[1])
    ax.set_title('annual anomaly '+fvar)
    plt.colorbar(pcol)

    toto_pert=dspert_month.mean(dim='lon')
    toto_ref=dsref_month.mean(dim='lon')
    toto_ano=toto_pert-toto_ref
    vmin=np.min(toto_ano[cvar].values)
    vmax=np.max(toto_ano[cvar].values)
    ax = fig.add_subplot(grid[ivar,1])
    pcol=ax.pcolormesh(toto_ref['time'],toto_ref['lat'],toto_ano[cvar].transpose(),vmax=vmax,vmin=vmin)
    ax.contour(toto_ref['time'],toto_ref['lat'],toto_ano[cvar].transpose(),levels=[-99999,0,9999999],colors='w')
    date_form = DateFormatter("%m")
    ax.set_title('zonal monthly anomaly '+fvar)
    ax.xaxis.set_major_formatter(date_form)
    plt.colorbar(pcol)

    ax = fig.add_subplot(grid[ivar,2])
    ax.plot(dsref_ts['time'],dsref_ts[cvar])
    ax.plot(dspert_ts['time'],dspert_ts[cvar])
    ax.grid()
    ax.set_title('-80 to -60 mean '+fvar)
    date_form = DateFormatter("%m")
    ax.xaxis.set_major_formatter(date_form)
    
plt.savefig('forcing_perturbation_IPSLCM6_ssp585.png', format='png', dpi=150)
```
